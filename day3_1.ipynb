```python
import numpy as np

#1
# create array by numpy
A = np.array([[ 3, 5],[7, 5]])

# Requirements
B = np.array([[6],
             [9]])
X = np.linalg.inv(A).dot(B)
print(X)
# Plug in results
x = X[0][0]
y = X[1][0]
#check results 
print(int(3*x + 5*y))
print(int(7*x + 5*y))
```

```python
#2
# create array by numpy
A = np.array([[1,1,1],[0,1,1],[2,5,-1]])

# Requirements
B = np.array([[6],
             [-4],
              [27]])
X = np.linalg.inv(A).dot(B)
print(X)
# Plug in results
x = X[0][0]
y = X[1][0]
z = X[2][0]
#check results 
print(int(x + y+z))
print(int(y+ z))
print(int(2*x+5*y-z))
```
